## Упражнения по библиотеке Numpy

```python
import numpy as np
import random

def random_matrix(dim1 = 1, dim2 = 1, alen = 10, imin = 1, imax = 4):
    new_matrix = np.array([random.randrange(imin, imax) for _ in range(alen)])
    
    if (dim1 != 1 or dim2 != 1):
        return new_matrix.reshape(dim1, dim2)
    else:
        return new_matrix
```

**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
np.random.shuffle(random_array:=np.arange(4))
random_array[(random_array >= 3) & (random_array <= 8)] *= -1;
print(random_array)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
random_array[random_array.argmax()] = 0
print(random_array)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
print(np.stack(np.meshgrid(*(matrix := random_array.reshape(-1, 2)), indexing='ij'), axis=-1).reshape(-1, matrix.shape[0]))
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
np.random.shuffle(A:= np.arange(24).reshape(8, 3))
np.random.shuffle(B:= np.arange(4).reshape(2,2))
print(A[np.isin(A, B[0]).any(axis=1) & np.isin(A, B[1]).any(axis=1)])
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
big_matrix = random_matrix(10, 3, 30, 1, 5)
print(big_matrix[(big_matrix != big_matrix[:, [0]]).any(axis = 1)])
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
two_dim_array = random_matrix(-1, 2, 10, imin=1, imax=3)
print(np.unique(two_dim_array, axis=0))
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
x, y = np.array([random_matrix(1, 1, 10, 0, 2), random_matrix(1, 1, 10, 0, 2)])
x2 = x.copy()
y2 = y.copy()
print(x)
print(y)

# 1 - no numpy
def partition(ArrayLike: object, left: int, right: int):
    pivot = ArrayLike[right]
    i = left - 1
    for j in range(left, right):
        if (ArrayLike[j] < pivot):
            i += 1
            ArrayLike[j], ArrayLike[i] = ArrayLike[i], ArrayLike[j]
    pr = i + 1
    ArrayLike[pr], ArrayLike[right] = ArrayLike[right], ArrayLike[pr]
    return pr
            

def quick_sort(ArrayLike: object, left: int, right: int):
    if (left >= right): return
    pr = partition(ArrayLike, left, right)
    quick_sort(ArrayLike, left, pr - 1)
    quick_sort(ArrayLike, pr + 1, right)


quick_sort(x, 0, len(x)-1)
quick_sort(y, 0, len(y)-1)
print((x == y).all())

# 2 - numpy
print((np.sort(x2) == np.sort(y2)).all())
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])

# 1 - no numpy
counts = {}
for v in x:
    counts[int(v)] = counts.get(int(v), 0)  + 1
print((np.array([k for k in counts.keys()]), np.array([v for v in counts.values()])))

# 2 - numpy
changes = np.r_[True, x[:-1] != x[1:], True]
result = (x[changes[:-1]], np.diff(np.where(changes)[0]))
print(result)
```
